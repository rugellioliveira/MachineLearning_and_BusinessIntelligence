**Conhecendo Machine Learning:**
Considerando dados histricos de clientes que realizaram compras em nossa empresa, faa o agrupamento (segmentao) dos clientes por similaridade de caractersticas em 3 grupos e gerar relatrio para o setor de marketing.

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

K-Means  um algoritmo de clusterizao (ou agrupamento) disponvel na biblioteca Scikit-Learn.

 um algoritmo de aprendizado no supervisionado (ou seja, que no precisa de inputs de confirmao externos) que avalia e clusteriza os dados de acordo com suas caractersticas.

Carregando os dados:

```python
df_dsa = pd.read_csv('dados_clientes.csv')
df_dsa.head()
```

Anlise exploratria:

```python
#Resumo estatstico
df_dsa[['idade','renda_anual','pontuacao_gastos']].describe()
```

Pr-processamento dos dados:

```python
#Cria o padronizador dos dados
padronizador = StandardScaler()
```

```python
#Aplica o padronizador somente nas colunas de interesse
dados_padronizados = padronizador.fit_transform(df_dsa[['idade','renda_anual','pontuacao_gastos']])
```

```python
#Visualiza os dados
print(dados_padronizados)
```

Construo do modelo de machine learning para segmentao de clientes:

```python
#Definimos o nmero de clusters (k)
k = 3
```

```python
#Criamos o modelo K-means
kmeans = KMeans(n_clusters = k)
```

```python
#Treinamento do modelo com os dados padronizados
kmeans.fit(dados_padronizados)
```

```python
#Atribumos os rtulos dos clusters aos clientes
df_dsa['cluster'] = kmeans.labels_
```

```python
#Exibe o resultado (10 primeiras linhas)
df_dsa.head(10)
```

```python
#Salva o resultado em arquivo .csv
df_dsa.to_csv('segmentos.csv', index = False)
```
